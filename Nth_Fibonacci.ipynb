## My solution

```python
def getFib(n):
  F0 = 0
  F1 = 1
  FibSe = [F0]

  while len(FibSe) < n:
    FibSe.append(F1)
    Fn = F0 + F1
    F0 = F1
    F1 = Fn
  return FibSe[n-1]

getFib(6)
```

## Solution 1 (Recursive)

```python
def getFib(n):
  if n == 2:
    return 1
  elif n == 1:
    return 0
  else:
    return getFib(n- 1) + getFib(n- 2)
  

getFib(6)
```

## Solution 2 (Hash table + recursive)

```python
def getFib(n, memoize = {1:0, 2: 1}):
  if n in memoize:
    return memoize[n]
  else:
    memoize[n] = getFib(n- 1, memoize) + getFib(n- 2, memoize)
    return memoize[n]

getFib(6)
```

**O(n)** -> Time

**O(n)** -> Space

## Solution 2 (Iterative)

```python
def getFib(n):
  lastTwo = [0, 1]
  counter = 3
  while counter <= n:
    nextFib = lastTwo[0] + lastTwo[1]
    lastTwo[0] = lastTwo[1]
    lastTwo[1] = nextFib
    counter += 1
  return lastTwo[1] if n > 1 else lastTwo[0]


getFib(6)
```

**O(n)** -> Time

**O(1)** -> Space
